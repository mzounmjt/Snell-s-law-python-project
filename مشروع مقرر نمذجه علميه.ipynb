```python
import math
def Refraction_angles(data):
    angles_of_incidence = data["Angle incidence (Î¸1)"]
    refractive_m1 = data['n1']
    refractive_m2 = data['n2']
    refraction_angles = []
    for i in range(len(angles_of_incidence)):
        angle_incidence = angles_of_incidence[i]
        refractive_index_m1 = refractive_m1[i]
        refractive_index_m2 = refractive_m2[i]
        angle_of_incidence= math.radians(angle_incidence)
        sin_angle_of_refraction = (refractive_index_m1 / refractive_index_m2) * math.sin(angle_of_incidence)
        angle_of_refraction = math.degrees(math.asin(sin_angle_of_refraction))
        refraction_angles.append(angle_of_refraction)
    return refraction_angles
data = {
    "Angle incidence (Î¸1)": [30, 40, 50, 60, 70],
    'n1': [1, 1, 1, 1, 1],
    'n2': [1.46, 1.52, 1.53, 1.47, 1.46]}
result_angles = Refraction_angles(data)
print("Angles of Refraction:" ,result_angles)
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np
n1 = 1    #air
n2 = 1.52 #glass
theta1_deg = 40
theta1_rad = math.radians(theta1_deg)
theta2_rad = math.asin(n1 * math.sin(theta1_rad) / n2)
theta2_deg = math.degrees(theta2_rad)
boundary_x = np.linspace(-1, 1, 400)
boundary_y = np.zeros_like(boundary_x)
#incident ray
incident_x = np.linspace(-0.75, 0, 400)
incident_y = incident_x * np.tan(theta1_rad)
#Refracted_ray
refracted_x = np.linspace(0, 0.75, 400)
refracted_y = refracted_x * np.tan(theta2_rad)
#Light Path throgh two mediums Graph
plt.figure(figsize=(8, 8))
plt.plot(boundary_x, boundary_y, label='Boundary (air/plate glass)')
plt.plot(incident_x, incident_y, label='Incident ray')
plt.plot(refracted_x, refracted_y, label='Refracted ray')
plt.legend()
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Light Refraction from Air to glass ')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
